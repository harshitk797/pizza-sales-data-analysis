```python
#importing the required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# importing the csv dataset 

df = pd.read_csv(r"C:\Users\harsh\OneDrive\Desktop\SpaceCE\DataSet.csv")

df.head()
```

```python
# General overview of data
df.info()
```

```python
#filling 0 to the empty columns

df["quantity"].fillna(0,inplace=True)
df["unit_price"].fillna(0,inplace=True)
df["total_price"].fillna(0,inplace=True)
df.info()
```

```python
# Changing from float to integer
df["quantity"] = df["quantity"].astype(int)
df.info()
```

```python
# Describing the data
df.describe()
```

##### Data Cleaning 
### Checking for duplicates

```python
# Using Conditional Statements

if df.duplicated().sum() > 0:
    print("Duplicates are present")
else:
    print("No duplicates exist")
```

#### Exploratory Data Analysis
## Daily Sales Trends
## Pizza Category Analysis

```python
df.head()
```

```python
daily_sales = df.groupby('order_date')['total_price'].sum()
plt.figure(figsize=(14,5))
daily_sales.plot()
plt.title("Daily Sales Trend")
plt.xlabel("Date")
plt.ylabel("Sales")
plt.show()
```

### Pizza Category Analysis

```python
sns.countplot(data=df,x='pizza_category')
plt.title("Orders by Pizza Category")
plt.show()
```

### Top 10 Best Selling Pizza

```python
top=df.groupby('pizza_name')['quantity'].sum().sort_values(ascending=True).head(10)

plt.figure(figsize=(10,5))
top.plot(kind='barh')
plt.title("Top 10 Best Selling Pizzas")
plt.show()
```

### Top Revenue Generating Pizza

```python
top_rev=df.groupby('pizza_name')['total_price'].sum().sort_values(ascending=True).head(10)

top_rev.plot(kind='barh')
```

### Revenue Distribution

```python
sns.histplot(df['total_price'],bins=30,kde=True)
```

#### Hourly Orders

```python
df['hour']=pd.to_datetime(df['order_time'].astype(str)).dt.hour

sns.countplot(x='hour',data=df)
plt.ylabel('order_count')
plt.title("Orders by Hour")
plt.show()
```
